```python
import pandas as pd
```

```python
# Read in influencer
influencer = pd.read_csv('/Users/katesxiao/Documents/M19/IS/Data/combined_average_maximum_influence.csv',lineterminator='\n')
influencer.head()
```

```python
# format the table
influencer.rename(columns={'timestamp':'Date'}, inplace = True)
influencer['Date']=pd.to_datetime(influencer['Date'])
influencer.head()
```

```python
# Read in Bitcoin price
headers = ['Unix Timestamp','Date','Symbol','Open','High','Low','Close','Volume']
price_1hr = pd.read_csv ('/Users/katesxiao/Documents/M19/IS/Data/raw_bitcoin.csv',names=headers)
price_1hr=price_1hr.drop([0,1,2])
price_1hr.tail()
```

```python
# format Date column
price_1hr['Date']=pd.to_datetime(price_1hr['Date'],dayfirst=True)
price_1hr.tail()
```

```python
# price table left outer join influencer table
merge_price_influencer = pd.merge(price_1hr, influencer, on='Date', how='left')
merge_price_influencer.head()
```

```python
influencer.shape
```

```python
# non-empty records
merge_notnull = merge_price_influencer.query('Average_Anger == Average_Anger')
merge_notnull
```

```python
merge_notnull.shape
```

```python
# fill NaN with 0
merge_null2zero = merge_price_influencer.fillna(0)
merge_null2zero
```

```python
# query non-empty records
merge_price_influencer.query('Average_Anger>0')
```

```python
# output merged table
merge_null2zero.to_csv(r'/Users/katesxiao/Documents/M19/IS/Data/combined_average_maximum_influence_null.csv', index = False)
```
