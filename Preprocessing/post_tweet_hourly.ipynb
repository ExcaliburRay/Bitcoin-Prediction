```python
import pandas as pd
tweet = pd.read_csv('/Users/katesxiao/Documents/M19/IS/Data/tweet_average_maximum_influence.csv')
post = pd.read_csv('/Users/katesxiao/Documents/M19/IS/Data/post_average_maximum_influence.csv')
tweet = tweet.loc[:, ~tweet.columns.str.contains('^Unnamed')]
post = post.loc[:, ~post.columns.str.contains('^Unnamed')]

t = tweet.set_index('tweetTimestamp')
p = post.set_index('postTimestamp')
```

```python
new_header = ['Average_Anger','Average_Joy','Average_Sadness',
              'Average_Fear','Average_Disgust','Maximum_Anger',
              'Maximum_Joy','Maximum_Sadness','Maximum_Fear',
              'Maximum_Disgust']
t.columns = new_header
p.columns = new_header
```

```python
result = pd.concat((t, p))
result = result.groupby(result.index).mean()
result
```
