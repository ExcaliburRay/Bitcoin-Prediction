```python
import pandas as pd

reddit = pd.read_csv('/Users/katesxiao/Documents/M19/IS/Data/reddit_influential.csv')
reddit
```

```python
post = reddit[['post_id','weighted_anger','weighted_joy','weighted_sadness','weighted_fear',
                 'weighted_disgust','timestamp']]
new_header = ['id','Anger','Joy','Sadness','Fear','Disgust','timestamp']
post.columns = new_header
post = post.set_index('id')
post.reset_index(drop = True)
post
```

```python
post_average = post.groupby('timestamp', as_index = False)['Anger','Joy','Sadness','Fear','Disgust'].mean()
new_header = ['postTimestamp','postAverage_Anger','postAverage_Joy','postAverage_Sadness',
              'postAverage_Fear','postAverage_Disgust']
post_average.columns = new_header
```

```python
post_maximum = post.groupby('timestamp', as_index = False)['Anger','Joy','Sadness','Fear','Disgust'].max()
new_header = ['timestamp2','postMaximum_Anger','postMaximum_Joy','postMaximum_Sadness',
              'postMaximum_Fear','postMaximum_Disgust']
post_maximum.columns = new_header
```

```python
post_result = pd.concat([post_average, post_maximum], axis = 1, sort = False)
post_result.drop("timestamp2",axis = 1, inplace = True)
post_result.set_index('postTimestamp')
post_result
```

```python
post_result.to_csv(r'/Users/katesxiao/Documents/M19/IS/Data/post_average_maximum_influence.csv')
```
