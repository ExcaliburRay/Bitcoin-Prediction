```python
# read all aip output file
import pandas as pd
import glob

path = r'/Users/katesxiao/Documents/M19/IS/Data/post_api_output'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

post_api = pd.concat(li, axis=0, ignore_index=True)
post_api
```

```python
# get header to output
new_header = ['id','Anger','Joy','Sadness','Fear','Disgust']
post_api.columns = new_header
post_api = post_api.set_index('id')
post_api
```

```python
# read the entire input file
post_processed = pd.read_csv('/Users/katesxiao/Documents/M19/IS/Data/processed_reddit_post.csv')
post_processed = post_processed[['post_id','timestamp']]
post_processed = post_processed.set_index('post_id')
post_processed
```

```python
# check the shape
# line number of these two files should be equal
post_api.shape
```

```python
post_processed.shape
```

```python
post = pd.concat([post_api, post_processed], axis=1, sort = False)
post
```

```python
post.reset_index(drop = True)
```

```python
post_average = post.groupby('timestamp', as_index = False)['Anger','Joy','Sadness','Fear','Disgust'].mean()
new_header = ['postTimestamp','postAverage_Anger','postAverage_Joy','postAverage_Sadness',
              'postAverage_Fear','postAverage_Disgust']
post_average.columns = new_header
```

```python
post_maximum = post.groupby('timestamp', as_index = False)['Anger','Joy','Sadness','Fear','Disgust'].max()
new_header = ['timestamp2','postMaximum_Anger','postMaximum_Joy','postMaximum_Sadness',
              'postMaximum_Fear','postMaximum_Disgust']
post_maximum.columns = new_header
```

```python
post_result = pd.concat([post_average, post_maximum], axis = 1, sort = False)
post_result.drop("timestamp2",axis = 1, inplace = True)
post_result.set_index('postTimestamp')
post_result
```

```python
post_result.to_csv(r'/Users/katesxiao/Documents/M19/IS/Data/post_average_maximum.csv')
```
