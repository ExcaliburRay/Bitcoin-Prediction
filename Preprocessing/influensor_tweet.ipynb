```python
import pandas as pd
tweets = pd.read_csv('/Users/katesxiao/Documents/M19/IS/Data/influencer_input_tweet.csv', lineterminator = '\n')
bitcoin = pd.read_csv('/Users/katesxiao/Documents/M19/IS/Data/influencer_input_bitcoin.csv')
```

```python
tweets['Sentiment_Value'] = tweets[['Anger','Joy','Sadness','Fear','Disgust']].max(axis=1)

sentiment_tag = list(tweets[['Anger','Joy','Sadness','Fear','Disgust']].idxmax(axis = 1))
li = []
for i in sentiment_tag:
    if i == 'Joy':
        li.append('Up')
    else:
        li.append('Down')
tweets['sentiment'] = sentiment_tag
tweets['movement'] = li
tweets.head()
```

```python
def getBoolean(li, threshold):
    result = []
    for i in li:
        if i > threshold:
            result.append(1)
        else:
            result.append(0)
    return result
```

```python
# average of the entire columns over all tweets
likes_average = tweets.loc[:,'likes'].mean()
replies_average = tweets.loc[:,'replies'].mean()
retweets_average = tweets.loc[:,'retweets'].mean()

# take the column out
li_likes = list(tweets['likes'])
li_replies = list(tweets['replies'])
li_retweets = list(tweets['retweets'])

# get the result as a 0 and 1 boolean list and get columns that have larger than average
result_likes = getBoolean(li_likes, likes_average)
result_replies = getBoolean(li_replies, replies_average)
result_retweets = getBoolean(li_retweets, retweets_average)

result = []
for i in range(len(result_likes)):
    total = result_likes[i] + result_replies[i] + result_retweets[i]
    result.append(total)
tweets['weight'] = result
```

```python
user = tweets.groupby(['user']).count()
print(user.shape)
user = user[user.timestamp != 1]
user = user.reset_index()
user_more_than_one = list(user['user'])
print(len(user_more_than_one))
```

```python
tweets_influ = tweets[tweets.user.isin(user_more_than_one)]
tweets_influ = tweets_influ[tweets_influ.weight.isin([2,3])]
tweets_influ = tweets_influ.reset_index()
tweets_influ.shape
```

```python
index_over_threshold = list(tweets_influ['index'])

price = bitcoin['Close'].values.tolist()

BT_timestamp = bitcoin['Date'].values.tolist()
post_timestamp = tweets['timestamp'].values.tolist()

# find the actual average bitcoin price change
price_actual_trend = []
count_valueError = 0
for i in index_over_threshold:
    sum_price_end = 0
    index_begin = BT_timestamp.index(post_timestamp[i])
    price_begin = float(price[index_begin])
    print("-------Index is %d-------" % (i))
    for j in range(1,52):
        index_end = index_begin + j
        sum_price_end += float(price[index_end])
    price_end = sum_price_end/51
    if price_end >= price_begin:
        price_actual_trend.append('Up')
    else:
        price_actual_trend.append('Down')
```

```python
tweets_influ['real_movement'] = price_actual_trend

li = []
for i in range(len(tweets_influ['movement'])):
    if tweets_influ['movement'][i] == tweets_influ['real_movement'][i]:
        li.append('T')
    else:
        li.append('F')
tweets_influ['Accuracy'] = li
tweets_influ
```

```python
temp_df = tweets_influ.groupby(['user']).Accuracy.value_counts().unstack(fill_value = 0)
temp_df_two = tweets_influ.groupby(['user']).mean()
```
