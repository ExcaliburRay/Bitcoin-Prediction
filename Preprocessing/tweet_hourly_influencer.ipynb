```python
def key_func(x):
        return os.path.split(x)[-1]
```

```python
# read all aip output file
import pandas as pd
import glob
import os

path = r'/Users/katesxiao/Documents/M19/IS/Data/tweet_api_output'
all_files = sorted(glob.glob(path + "/*.csv"))

li = []

for filename in sorted(glob.glob(path + "/*.csv"), key = key_func):
# for filename in sorted(os.listdir(path)):
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
    print("Current File Being Processed is: " + filename)

tweet_api = pd.concat(li, axis=0, ignore_index=True)
tweet_api
```

```python
# get header to output
new_header = ['id','Anger','Joy','Sadness','Fear','Disgust']
tweet_api.columns = new_header
tweet_api
```

```python
# read the entire input file
tweet_eng = pd.read_csv('/Users/katesxiao/Documents/M19/IS/Data/tweet_eng.csv',
                             lineterminator = '\n')
tweet_eng = tweet_eng[['user','tweet-id','timestamp']]

# round the time to hourly
tweet_eng['timestamp']=pd.to_datetime(tweet_eng['timestamp'])
tweet_eng['timestamp']=pd.Series(tweet_eng['timestamp']).dt.round("H")
tweet_eng
```

```python
tweet = pd.concat([tweet_api, tweet_eng], axis=1, sort = False)
tweet
```

```python
influence = pd.read_csv('/Users/katesxiao/Documents/M19/IS/Data/influential_tweet_user.csv')
all_users = list(influence['user'])
all_weights = list(influence['user_weight'])
```

```python
influence = influence.set_index('user')
influence
```

```python
%%time
for i in range(len(tweet['user'])):
    if tweet['user'][i] in all_users:
        tweet['Anger'][i] = tweet['Anger'][i] * influence.loc[tweet['user'][i],'user_weight']
        tweet['Joy'][i] = tweet['Joy'][i] * influence.loc[tweet['user'][i],'user_weight']
        tweet['Sadness'][i] = tweet['Sadness'][i] * influence.loc[tweet['user'][i],'user_weight']
        tweet['Fear'][i] = tweet['Fear'][i] * influence.loc[tweet['user'][i],'user_weight']
        tweet['Disgust'][i] = tweet['Disgust'][i] * influence.loc[tweet['user'][i],'user_weight']
        print("Current line is %d" % (i))
        
```

```python
tweet
```

```python
tweet_average = tweet.groupby('timestamp', as_index = False)['Anger','Joy','Sadness','Fear','Disgust'].mean()
new_header = ['tweetTimestamp','tweetAverage_Anger','tweetAverage_Joy','tweetAverage_Sadness',
              'tweetAverage_Fear','tweetAverage_Disgust']
tweet_average.columns = new_header
tweet_average.shape
```

```python
tweet_maximum = tweet.groupby('timestamp', as_index = False)['Anger','Joy','Sadness','Fear','Disgust'].max()
new_header = ['timestamp2','tweetMaximum_Anger','tweetMaximum_Joy','tweetMaximum_Sadness',
              'tweetMaximum_Fear','tweetMaximum_Disgust']
tweet_maximum.columns = new_header
tweet_maximum.shape
```

```python
tweet_result = pd.concat([tweet_average, tweet_maximum], axis = 1, sort = False)
tweet_result.drop("timestamp2",axis = 1, inplace = True)
tweet_result.set_index('tweetTimestamp')
tweet_result
```
