```python
import pandas as pd
#make sure the sequence of price is correct
price_data = pd.read_csv('../Files/price_sentiment_zero_label.csv')
price_data
```

```python
price_data = price_data.iloc[:,[0,1,7,9,10,11,12,13,19,20]]
actual_label_list = list(price_data['label'])
print(len(actual_label_list))
# price_data['Unnamed: 0'] = list(price_data['Unnamed: 0'][::-1])
print(price_data.tail())
```

```python
import numpy as np
def create_dataset_ps(dataset,look_back,future_hours,price_column_index):
    dataX, dataY = [], []
    for i in range(look_back, len(dataset)):
        a = list(dataset[i-look_back : i, price_column_index])
        a.extend(list(dataset[i-look_back : i, 3]))
        a.extend(list(dataset[i-look_back : i, 4]))
        a.extend(list(dataset[i-look_back : i, 5]))
        a.extend(list(dataset[i-look_back : i, 6]))
        a.extend(list(dataset[i-look_back : i, 7]))
        dataX.append(a)
        if i+6 < len(dataset):
            dataY.append(dataset[i+ future_hours , price_column_index])
        else:
            dataY.append(0)
    print(len(dataY))
    return np.array(dataX), np.array(dataY)
```

```python
price_data_np = price_data.as_matrix()
future = 6
back = 51
price_column_index = 2
PX, PY = create_dataset_ps(price_data_np,back,future,price_column_index)
```

```python
# Use label to test accuracy
#print(PY_test.head())
#len(PSY_pred)
#xset is the dataframe of test training x
def test_label(xset,prdt_price_set,test_label_set):
    counter = 0
    result = []
    boolean_result = []
    label_0 = 0
    label_1  = 0
    label_2  = 0
    label_3 = 0
    result_label = []
    xset = xset.set_index(pd.Series(range(0,xset.shape[0])))
    prdt_price_set = prdt_price_set.set_index(pd.Series(range(0,prdt_price_set.shape[0])))
    #print(xset.head())
    #print(prdt_price_set.head())
    for i,row in xset.iterrows():
        #print(i)
        x = row[back-1]
        #print(str(i)+':  '+str(x))
        #print(type(prdt_price_set))
        y = prdt_price_set.iloc[i,0]
        percent = y/x-1
        #print(str(i)+':  '+str(y))
        actual_label = test_label_set[i]
        if percent>0.015:
            test_label_result = 0
            label_0 += 1
            result_label.append(test_label_result)
        elif percent>0:
            test_label_result = 1
            label_1 += 1
            result_label.append(test_label_result)
        elif percent > -0.015:
            test_label_result = 2
            label_2 += 1
            result_label.append(test_label_result)
        else:
            test_label_result = 3
            label_3 += 1
            result_label.append(test_label_result)
        result.append(test_label_result)
        if test_label_result == actual_label:
            counter += 1
            boolean_result.append(1)
        else:
            boolean_result.append(0)
    print('# of correct prediction: ' + str(counter))
    print('Size of test: '+ str(xset.shape[0]))
    print('Accuracy: '+str(counter/xset.shape[0]))
    
    label_distribution = [ label_0, label_1, label_2, label_3]
    return boolean_result, label_distribution, result_label
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
import matplotlib.pyplot as plt
def test_regression(cut_off,PX_dataset,PY_dataset,trun_actual_label_list):
    model = LinearRegression()
    PX_train = PX_dataset[0:int(len(PX_dataset)*cut_off)]
    PX_test = PX_dataset[int(len(PX_dataset)*cut_off):len(PX_dataset)]
    PY_train = PY_dataset[0:int(len(PY_dataset)*cut_off)]
    PY_test = PY_dataset[int(len(PY_dataset)*cut_off):len(PY_dataset)]
    PY_test_label = trun_actual_label_list[int(len(trun_actual_label_list)*cut_off):len(trun_actual_label_list)]
    model.fit(PX_train,PY_train)
    print(str(cut_off*100)+"%training")
    score = model.score(PX_test,PY_test)
    print(score)
    PSY_pred = model.predict(PX_test) 
    boolean_result, label_distribution, result_label = test_label(PX_test,PY_test,PY_test_label)
    return boolean_result, label_distribution, result_label, PY_test_label
    #plt.figure()
    #plt.plot(range(len(result_list[:200])),result_list[:200],'b',label="predict")
    #plt.legend(loc="upper right")
    #plt.show()
```

```python
boolean_result, label_distribution, result_label, PY_test_label = test_regression(0.6,PX,PY,trun_actual_label_list)
```
