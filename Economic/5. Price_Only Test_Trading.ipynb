```python
import pandas as pd
#make sure the sequence of price is correct
price_data = pd.read_csv('../Files/price_emotion_4label_6hour.csv')
price_data[1000:1050]
```

```python
price_data = price_data.iloc[:,[0,1,7,19,20]]
actual_label_list = list(price_data['label'])
actual_percentage_list = list(price_data['Percentage'])
print(len(actual_label_list))
print(len(actual_percentage_list))
price_data['Unnamed: 0'] = list(price_data['Unnamed: 0'][::-1])
print(price_data.tail())
```

```python
import numpy as np
def create_dataset(dataset,look_back,future_hours,price_column_index):
    dataX, dataY = [], []
    for i in range(look_back, len(dataset)):
        a = list(dataset[i-look_back : i, price_column_index])
        dataX.append(a)
        if i+6 < len(dataset):
            dataY.append(dataset[i+ future_hours , price_column_index])
        else:
            dataY.append(0)
    print(len(dataY))
    return np.array(dataX), np.array(dataY)
```

```python
price_data_np = price_data.as_matrix()
future = 6
back = 51
price_column_index = 2
PX, PY = create_dataset(price_data_np,back,future,price_column_index)
PX = pd.DataFrame(PX)
PY = pd.DataFrame(PY)
#truncate last 6 rows
trun_actual_label_list = actual_label_list[back:-future]
trun_actual_percentage_list = actual_percentage_list[back:-future]
PY = PY[:-future]
PX = PX[:-future]
print(len(trun_actual_label_list))
print(len(PY))
```

```python
PX.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
PX_train = PX[0:int(len(PX)*0.7)]
PX_test = PX[int(len(PX)*0.7):len(PX)]
PY_train = PY[0:int(len(PY)*0.7)]
PY_test = PY[int(len(PY)*0.7):len(PY)]
PY_test_label = trun_actual_label_list[int(len(trun_actual_label_list)*0.7):len(trun_actual_label_list)]
PY_test_percentage = trun_actual_percentage_list[int(len(trun_actual_label_list)*0.7):len(trun_actual_label_list)]
model.fit(PX_train,PY_train)
```

```python
# R² score of our model
score = model.score(PX_test,PY_test)
print(score)
```

```python
PSY_pred = model.predict(PX_test) 
```

```python
# Use label to test accuracy
#print(PY_test.head())
#len(PSY_pred)
initial = 100
#xset is the dataframe of test training x
def test_label(xset,prdt_price_set,test_label_set):
    counter = 0
    result = []
    boolean_result = []
    label_0 = 0
    label_1  = 0
    label_2  = 0
    label_3 = 0
    result_label = []
    xset = xset.set_index(pd.Series(range(0,xset.shape[0])))
    prdt_price_set = prdt_price_set.set_index(pd.Series(range(0,prdt_price_set.shape[0])))
    #print(xset.head())
    #print(prdt_price_set.head())
    for i,row in xset.iterrows():
        #print(i)
        x = row[back-1]
        #print(str(i)+':  '+str(x))
        #print(type(prdt_price_set))
        y = prdt_price_set.iloc[i,0]
        percent = y/x-1
        #print(str(i)+':  '+str(y))
        actual_label = test_label_set[i]
        if percent>0.015:
            test_label_result = 0
            label_0 += 1
            result_label.append(test_label_result)
        elif percent>0:
            test_label_result = 1
            label_1 += 1
            result_label.append(test_label_result)
        elif percent > -0.015:
            test_label_result = 2
            label_2 += 1
            result_label.append(test_label_result)
        else:
            test_label_result = 3
            label_3 += 1
            result_label.append(test_label_result)
        result.append(test_label_result)
        if test_label_result == actual_label:
            counter += 1
            boolean_result.append(1)
        else:
            boolean_result.append(0)
    print('# of correct prediction: ' + str(counter))
    print('Size of test: '+ str(xset.shape[0]))
    print('Accuracy: '+str(counter/xset.shape[0]))
    label_distribution = [ label_0, label_1, label_2, label_3] #predicted one
    return boolean_result, label_distribution, result_label#real one
```

```python
boolean_result, label_distribution, result_label = test_label(PX_test,PY_test,PY_test_label)
#boolean: 1,0 true or false
#label_distr: four buckets result(X,X,X,X)
#result_label : predicted labels
```
