# This code is to split the buckets based on 6 hours
* Split the bucket into >1.5%, 0%-1.5%, -1.5%-0%, <-1.5%
* The percentage is calculated by the amount of (24hr_latter_price - current_price)/current_price
* Labeled each row from Class 0 to Class 3

```python
import pandas as pd
data2 =  pd.read_csv ('../Files/combined_average_maximum_influence_null.csv')
data2['Date']=pd.to_datetime(data2['Date'])
data2 = data2.sort_values(['Date'], ascending=[True])
data2.head()
```

### Calculate the percentage difference between time t and t+6 hr

```python
percentage_6hour = []
for i in range(data2.shape[0]-1, 5, -1):
  price_now = data2.loc[i]['Close']
  price_6hours = data2.loc[i-6]['Close']
  percentage_6hour.append((price_6hours-price_now)/price_now)

for i in range(6):
  percentage_6hour.append(0)
```

```python
data2['Percentage']=percentage_6hour
data2 = data2[:data2.shape[0]-6]
```

```python
data2.head()
```

```python
data2.to_csv('../Files/6hour_price_emotion_new.csv', encoding='utf-8', index=False)
data3 = pd.read_csv ('../Files/6hour_price_emotion_new.csv')
```

```python
data3.head()
```

# Labeled each row into different buckets: 

```python
# Split buckets
up1 = 0.015; down1 = -0.015;

bucket1 = 0; bucket2 = 1;  bucket3 = 2; bucket4 = 3; 
count1 = 0; count2 = 0;   count3 = 0; count4 = 0;

label = []

for i in range(data3.shape[0]):
  
  p = data3.loc[i]['Percentage']
  
  if (p > up1):
    label.append(bucket1)
    count1 = count1+1
  elif (0 <= p and p <= up1):
    label.append(bucket2)
    count2 = count2+1
  elif (down1 <= p and p < 0):
    label.append(bucket3)
    count3 = count3+1
  elif (p<down1):
    label.append(bucket4)
    count4 = count4+1
       
print(len(label))
```

```python
print(count1)
print(count2)
print(count3)
print(count4)
```

```python
data3['label']=label
data3.head()
```

```python
data3.to_csv('../Files/6hour_price_emotion_4label.csv', encoding='utf-8', index=False)
data4 = pd.read_csv ('../Files/6hour_price_emotion_4label.csv')
```
