```python
import pandas as pd
price_emotion = pd.read_csv('../Files/price_emotion_4label_6hour.csv')
price_emotion.head()
```

```python
# price_emotion = price_emotion.iloc[::-1]
price_emotion.index = range(1,len(price_emotion) + 1)
price_emotion.drop(price_emotion.columns[0], axis=1,inplace=True)
price_emotion
```

```python
# how many rows before the first nonzero
count=0
for i in range(1,len(price_emotion)+1,+1):
  if (price_emotion.loc[i]['Average_Anger'] == 0.0 and price_emotion.loc[i]['Average_Joy'] == 0.0 and price_emotion.loc[i]['Average_Sadness'] == 0.0 and price_emotion.loc[i]['Average_Fear'] == 0.0 and price_emotion.loc[i]['Average_Disgust'] == 0.0):
    count = count + 1;
  else:
    print("First row of nonzero value: " + str(i));
    break;
print(count);
```

```python
# how many rows after the last nonzero
count2=0
for j in range(len(price_emotion),-1,-1):
    if (price_emotion.loc[j]['Average_Anger'] == 0.0 and price_emotion.loc[j]['Average_Joy'] == 0.0 and price_emotion.loc[j]['Average_Sadness'] == 0.0 and price_emotion.loc[j]['Average_Fear'] == 0.0 and price_emotion.loc[j]['Average_Disgust'] == 0.0):
    count2 = count2 + 1
    else:
    print("First row of nonzero value: " + str(j));
    break;
print(count2)  
```
